```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution of a numerical column
sns.histplot(df['annual_income'], bins=30)
plt.show()

# Countplot for a categorical column
sns.countplot(data=df, x='loan_status')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='grade', y='loan_amount')
plt.title('Loan Amount Distribution by Grade')
plt.show()
```

relationship between income and loan amount?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='annual_income', y='loan_amount', hue='loan_status', alpha=0.6)
plt.title('Annual Income vs. Loan Amount by Loan Status')
plt.show()
```

To see loan trends over time

```python
# Convert issue_date to datetime if not already
df['issue_date'] = pd.to_datetime(df['issue_date'], errors='coerce')

# Group by date and calculate mean loan amount
loan_trend = df.groupby(df['issue_date'].dt.to_period("M"))['loan_amount'].mean()

plt.figure(figsize=(12, 6))
loan_trend.plot()
plt.title('Average Loan Amount Over Time')
plt.xlabel('Issue Date')
plt.ylabel('Average Loan Amount')
plt.show()
```
